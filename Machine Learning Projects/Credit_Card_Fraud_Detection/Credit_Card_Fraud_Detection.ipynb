```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset into pandas DataFrame
data = pd.read_csv('creditcard.csv')
```

```python
data
```

```python
# Dataset Information
data.info()
```

```python
# Distribution of legit transactions and fraudulent transactions
data['Class'].value_counts()
```

```python
# Seperating the data for analysis
legit = data.loc[data['Class'] == 0]
fraud = data.loc[data['Class'] == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
data.groupby('Class').mean()
```

```python
# Building a sample dataset containing similar distribution of Normal Transactions and Fraudulent transactions
legit_sample = legit.sample(n=241)
```

```python
# Concatenating two DataFrames
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data
```

```python
new_data.Class.value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# Splitting the data into Features and Targets
x = new_data.drop(['Class'], axis=1)
y = new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting the data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
# Creating the Model
model = LogisticRegression()
```

```python
# Model Training
model.fit(x_train, y_train)
```

```python
# Model Evaluation

# Prediction on training data
prediction_on_training_data = model.predict(x_train)

training_accuracy = accuracy_score(y_train, prediction_on_training_data)
print(f'Accuracy on Training data = {training_accuracy}')


# Prediction on test data
prediction_on_test_data = model.predict(x_test)

test_accuracy = accuracy_score(y_test, prediction_on_test_data)
print(f'Accuracy on test data = {test_accuracy}')
```
